```python
import pandas as pd
import re
```

```python
df = pd.read_excel("/content/drive/MyDrive/Bangla Hate Speech Detection/dataset/RawData.xlsx")
df.head(5) 
```

```python
def punctuations_remover(my_str):
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~।'''
    no_punct = " "
    for char in my_str:
        if char not in punctuations:
            no_punct = no_punct + char
    no_punct = no_punct.replace("\n",' ')
    return no_punct
```

```python
df.comment = df.comment.apply(lambda x: punctuations_remover(x))
```

```python
######### foreign language removing #######
def for_lan_re(text):
    text = "".join(i for i in text if i in [".","।"] or 2432 <= ord(i) <= 2559 or ord(i)== 32)
    sentence = re.sub(' +', ' ', text)
    return sentence
```

```python
df.comment = df.comment.apply(lambda x: for_lan_re(x))
```

```python
from bnlp.corpus import stopwords
from bnlp.corpus.util import remove_stopwords
```

```python
def stopword_removing(raw_text):
    result = remove_stopwords(raw_text, stopwords)
    return ' '.join(result)
```

```python
df['withoutStopwordComment'] =  df.comment.apply( lambda x : stopword_removing(x))
```

```python
df.head()
```

```python
df = df[["comment","withoutStopwordComment","Sentiment"]]
```

```python
df.head()
```

```python
df['category'] = df.Sentiment.replace(['Hate Speech','Normal Speech'],[0,1])
```

```python
## dataset balancing ##
```

```python
import pandas as pd
import matplotlib.pyplot as plt
value_counts = df['category'].value_counts()

# Plot a pie chart
value_counts.plot.pie(autopct='%.1f%%')

# Add a title
plt.title("Sample Pie Chart")

# Show the plot
plt.savefig("Imbalanced Dataset.jpeg", dpi = 250)
plt.show()
```

```python
# count the number of samples in each class
class_counts = df['category'].value_counts()
print(class_counts)
```

```python
# identify the class with the least samples
min_class = class_counts.idxmin()
max_class = class_counts.idxmax()
# calculate the number of samples required to balance the classes
upsample_size = class_counts[max_class] - class_counts[min_class]
print(upsample_size)
```

```python
## applying upsampling. 
# upsample the minority class to balance the classes
# It samples the data from the data frame with replacement, meaning that it can choose the same sample multiple times.
df_minority_upsampled = df[df['category'] == min_class].sample(n=upsample_size, replace=True, random_state=42)
df_upsampled = pd.concat([df, df_minority_upsampled])
# shuffle the samples in the balanced dataset
df_upsampled = df_upsampled.sample(frac=1, random_state=42).reset_index(drop=True)
# save the balanced dataset to a file
```

```python
import pandas as pd
import matplotlib.pyplot as plt
value_counts = df_upsampled['category'].value_counts()

# Plot a pie chart
value_counts.plot.pie(autopct='%.1f%%')

# Add a title
plt.title("Sample Pie Chart")

# Show the plot
plt.savefig("Balanced Dataset.jpeg", dpi = 250)
plt.show()
```
